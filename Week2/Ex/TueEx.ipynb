```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

```python
#타이타닉호 승객의 평균 나이를 구하라.
t1 = titanic['age'].mean()
t1
```

```python
#Ref
titanic.age.mean()  #괄호대신 .age로 접근이 가능하다.
```

```python
#타이타닉호 승객 중 여성 승객의 평균 나이를 구하라.
EqualFemale = titanic[titanic['sex'] == 'female']
FemaleAgeMean = EqualFemale['age'].mean()
FemaleAgeMean
```

```python
#Ref
titanic.loc[titanic.sex == 'female']['age'].mean()
```

```python
#Ref2
titanic.loc[titanic.sex == 'female'].age.mean()
```

```python
#Ref3
np.mean(titanic[titanic['sex'] == 'female']['age']).round(2)
```

```python
#타이타닉호 승객 중 1등실 선실을 탄 여성 승객의 평균 나이를 구하라.
EqualFirstFemale = titanic[(titanic['sex'] == 'female') & (titanic['class'] == 'First')]
FirstFemaleAgeMean = EqualFirstFemale['age'].mean()
FirstFemaleAgeMean 
```

```python
#Ref
titanic.loc[(titanic['class'] == 'First') & (titanic.sex == 'female')]['age'].mean()
```

```python
#category1 이라는 열 생성 20살보다 크거나 같으면 성별출력 하고 낮으면 child출력
titanic['category1'] = titanic.apply(lambda r : r.sex if r.age >= 20 else 'child', axis = 1)
titanic
```

```python
#타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고쳐라
titanic.age = titanic.age.fillna(t1).astype(int)    #방금 전 예제에서 만들어둔 나이 평균 t1 사용
titanic
```

```python
#Ref
age_mean = titanic.age.mean().round(0)
titanic['age'] = titanic.apply(lambda x : age_mean if pd.isna(x.age) else x.age, axis = 1).astype('i')  # ('i') == (int) 형으로 변환
titanic.tail(30)
```

```python
#타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category2열을 만들어라
# 성별을 나타내는 문자열 male 또는 female로 시작한다.
# 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열이 온다.
# 예른들어 27살 남성은 male27이 된다.

titanic['category2'] = titanic.sex+titanic.age.astype(str)
titanic
```

```python
#타이타닉호 승객을 '미성년자','청년','중년',장년','노년'나이 그룹으로 나눈다.
bins = [1, 20, 30, 50, 70, 100]
labels = ['미성년자', '청년', '중년', '장년', '노년']

TitanicCats = pd.cut(titanic.age, bins, labels=labels)
titanic['연령구분'] = pd.DataFrame(TitanicCats)
titanic
```

```python
# 각 나이 그룹의 승객 비율을 구한다. 비율의 전체 합은 1이 되어야 한다.
pd.DataFrame((titanic['연령구분'].value_counts()) / (titanic['연령구분'].value_counts().sum()))
```

```python
titanic.연령구분.value_counts()
```

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

titanic["연령대"] = pd.cut(titanic.age, bins, labels=labels) 
# 연령대 카테고리 추가

titanic_age = titanic.연령대.value_counts().to_frame()
titanic_age["비율"] = titanic_age.apply(lambda i : i["count"] / titanic_age["count"].sum(), axis=1) #비율 카테고리
titanic_age
```

```python
#타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category3 열을 만들어라..
# 20살 미만이면 성별에 관계없이 "미성년자" 라고 한다.
# 20살 이상이면 나이에 따라'청년','중년','장년','노년'을 구분하고 그뒤에 성별을 나타내는 '남성','여성'을 붙이다.
titanic['category3'] = titanic.apply(lambda i : i.연령구분+('남성' if i.sex == 'male' else '여성') if i.age >= 20 else '미성년자', axis=1).astype(str)
titanic
```
