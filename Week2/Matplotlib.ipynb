```python
# 선 종류 지정
#‘ - ‘ (Solid), ‘ - - ‘ (Dashed), ‘ : ‘ (Dotted), ‘ -. ‘ (Dash-dot)


plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')  #c0 파랑
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C1', label='Dashed') #0~9 는 각각 파란색, 주황색, 초록색, 빨간색, 보라색, 갈색, 분홍색, 회색, 노란색 순으로 지정되어 있음
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C2', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C3', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4, fontsize=8)
plt.show()
```

```python
#‘ - ‘ (Solid), ‘ - - ‘ (Dashed), ‘ : ‘ (Dotted), ‘ -. ‘ (Dash-dot)

plt.figure(figsize=(10,5 ))
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
```

```python
# 선 끝 모양 지정

plt.figure(figsize=(10, 6))
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', linewidth=5,
      solid_capstyle='butt', color='r', label='solid+butt')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='solid', linewidth=10,
      solid_capstyle='round', color='C0', label='solid+round')

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashed', linewidth=10,
      dash_capstyle='round', color='C1', label='dashed+round')
# dash_capstyle를 사용해서 선의 끝 모양 지정
# {'butt', 'projecting', 'round'}

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
#기본 사용 - fill_between() / 그래프 영역 채우기

x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)
#  (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역이 채워짐
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)
#  (x[2], y[2]), (x[3], y[3]), (0, y[2]), (0, y[3])을 잇는 영역이 채워짐

plt.show()
```

```python
# 두 그래프 사이 영역 채우기

x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[0:3], y1[0:3], y2[0:3], color='green', alpha=0.5)
# (x[1], y[1]), (x[1], y[2]), (x[2], y[1]), (x[2], y[2]) 사이 영역을 채움
plt.show()
```

```python
# 수평선/수직선 표시

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=4)
# axhline(y값, xmin, xmax) /xmin, xmax는 0에서 1사이 값. 0은 왼쪽 끝, 1은 오른쪽 끝
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)
# hlines(y, xmin, xmax) / 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시

plt.show()
```

```python
# 수직선 그리기

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
# axvline(x 값, ymin, ymax) 
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)
# vlines(x, ymin, ymax) / 점 (x, ymin)에서 점 (x, ymax)를 따라 수평선을 표시
plt.show()
```

```python
# Numpy 난수 분포 나타내기 

a = 2.0 * np.random.randn(10000) + 1.0
# 표준편차 2.0, 평균 1.0을 갖는 정규분포
b = np.random.standard_normal(10000)
# 표준정규분포
c = 20.0 * np.random.rand(5000) - 10.0
# 10.0에서 10.0 사이의 균일한 분포를 갖는 5000개의 임의의 값

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
# density=True로 설정해주면, 밀도함수가 되어서 막대의 아래 면적이 1이 됨
# alpha는 투명도

plt.show()
```

```python
#파이 차트
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%1.0f%%')
plt.show()
```

```python
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) #startangle의 기본값은 3시 수치로 각도를 조절하면 반시계 방향으로 시작점을 지정해줄 수 있다. 
```

```python
explode = [0,0,0,0.20]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=90,
        counterclock=False, explode=explode)
plt.show()
```

```python
# 중심에서 벗어나는 정도 설정 / 색상 지정

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]
# explode는 부채꼴이 파이 차트의 중심에서 벗어나는 정도를 설정
# Banana’와 ‘Grapes’ 영역에 대해서 반지름의 10% 만큼 벗어나도록 설정
colors = ['silver', 'gold', 'lightblue', 'lightgray']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
#shadow를 True로 설정하면, 파이 차트에 그림자가 표시
plt.show()
```

```python
# 부채꼴 스타일 지정

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.5, 'edgecolor': 'w', 'linewidth': 2}
#‘wedgeprops 딕셔너리의 ‘width’, ‘edgecolor’, ‘linewidth’ 키를 이용해서 각각 부채꼴 영역의 너비 (반지름에 대한 비율), 테두리의 색상, 테두리 선의 너비를 설정

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)

plt.show()
```
